# Data Processing

```python
# List of dataframes
citi_bike_data = [citi202203,citi202204,citi202205,citi202206,citi202207,citi202208,citi202209,citi202210,citi202211,citi202212, citi202301,citi202302, citi202303]

# Merge all dataframes into one
citi_bike_merge = pd.concat(citi_bike_data, axis=0, ignore_index=True)
citi_bike_merge.head()
```

```python
citi_bike_merge.isnull().sum()
```

```python
# Drop rows with missing values
citi_bike_cleaned = citi_bike_merge.dropna()

# Check the shape of the new dataframe
print("Shape of the original dataframe before dropping rows with missing values: ", citi_bike_merge.shape)
print("Shape of the new dataframe after dropping rows with missing values: ", citi_bike_cleaned.shape)
```

```python
citi_bike_cleaned.to_csv('citi_bike_cleaned.csv')
```
